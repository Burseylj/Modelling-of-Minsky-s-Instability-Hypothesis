# Introduction
Thirty years ago, economist Hyman Minsky argued that financial systems were "endogenously unstable,” and that their instability would increase over time (Minsky, 1986). He pointed to the increased interconnectedness of the global financial system, the creation of new nonbank financial entities, and the inability of financial regulators to keep up with increasingly complex financial instruments as issues that could lead to a financial crisis. As the reader in 2017 will know, those elements did in fact lead to the most significant financial and economic crisis since the Great Depression: the ‘Great Recession’ of 2007-09.

The stability of the financial system is important to everyone, from policymakers to pensioners. When will the next great crash occur? For this reason, there is great interest in modelling the economy.

The majority of economic models are static. These models, and the vast majority of economic thinkers, failed to predict the 'Great Recession.' This has led critics such as Australian economist Steve Keen to argue for the use of dynamic economic models instead. They argue that modelling the economy using differential equations allows for more complexity. Economists and mathematicians can examine the stability of such models of the world economy, perhaps giving insight into the operation of the highly complex real world version.

Minsky elaborated the ‘financial instability hypothesis,’ in which increased credit in the economy increases asset prices and the instability of the financial system. Keen (1995) combined Minsky's theories with economist Richard Goodwin's model of economic fluctuations and translated it into a mathematical model of the economy. This model was further expanded upon by Grasselli and Lima (2012).

Below we present three simplified models of the economy, as described by Grasselli and Lima (2012). The goal is to examine the equilibria and stability of the models, and see what they tell us about the model economy. 

# Goodwin Model (2D)

## Model Construction

### Model:

The first model is the 'Goodwin', a 2-dimensional economic growth model first elaborated by mathematician and economist Richard Goodwin. It is a form of the Lotka-Volterra model of predator-prey. The model is 

$$ \frac{d\omega}{dt}= \omega(f(\lambda)-\alpha) $$
$$ \frac{d\lambda}{dt}=\lambda \left(\frac{1-\omega}{v}-\delta-\alpha-\beta \right) $$

### Parameters:

#### The following are needed for the model derivation but do not appear in the model itself:

Let $Y$ be total yearly output in the economy (in dollars/year).

Let $K$ be the capital stock (in dollars).

Let $v$ be the capital-to-output ratio (constant by assumption) (in years).

Let $L$ be the number of employed persons (unit is people).

Let $a$ be labour productivity (units are dollars/person/year).

Let $N$ be the total labour force (unit is people).

Let $w$ be real wages per person (units are dollars/person/year).

#### The following parameters appear in the model (and are needed to derive the model from the economic assumptions):

Let $\alpha$ be the productivity growth rate, assumed constant (units are /year).

Let $\beta$ be the population growth rate, assumed constant (units are /year).

Let $\delta$ be the constant depreciation rate (the rate at which capital is used up and needs to be replaced) (units are /year, need $0 \leq \delta \leq 1 $ ).

Let $f(\lambda)$ be the Phillips curve. It is an increasing function with a vertical asymptote at $\lambda=1$, and has units of /year.

All quantities are in real and net terms. This means inflation is already accounted for, and intermediate revenues and expenditures are not included in the final output variables.

### Variables:

Let $\lambda$ be the employment rate, the number of employed persons divided by the total labour force ($\lambda=\frac{L}{N}$) (unitless, need $0 \leq \lambda \leq 1 $ for this to make sense).

Let $\omega$ be the share of the economy paid to labour through wages (unitless, need $0 \leq \omega \leq 1 $ for Goodwin model).

### Model derivation:

#### Assumptions:

The following economic assumptions are made, as in Grasselli and Lima (2012):

(1) Production depends only on two factors of production, capital and labour, both homogeneous (meaning that increasing the amount of inputs increases the output by the same proportion) (Goodwin, 1967, pg. 54). Grasselli and Lima (2012) assume a Leontief production function (fixed proportions production function).

$$ Y(t)= min [\frac{K(t)}{v},a(t)L(t)] $$

(2) The labour force is assumed to grow at a constant rate $\beta>0$. 

$$ N(t)=N_0e^{\beta t} $$

This means $\frac{dN}{dt} = \beta N $

(3) The employment rate is 

$$ \lambda(t)=\frac{L(t)}{N(t)} $$

(4) It is also assumed that labour productivity grows at a constant rate, $\alpha>0$.

$$ a(t)=a_0e^{\alpha t} $$

This means $\frac{da}{dt}=\alpha a $

(5) Full capital utilization is assumed, so production is equal to both the capital stock divided by the constant capital-to-output ratio, and labour productivity times the number of employed persons.

$$ \frac{K(t)}{v}=Y(t)=a(t)L(t) $$

(6) One of Goodwin's key assumptions is that the real wage rate "rises in the neighbourhood of full employment," ie it is an increasing function of the employment rate (Goodwin, pg. 54). This means that workers demand a higher real wage when the employment rate is higher, and accept lower wages when the employment rate is lower. 

Goodwin assumes that

$$ \frac{dw}{dt}=w(f(\lambda)) $$

where $f(\lambda)$ is an increasing function of the employment rate. In economic literature, this relationship is known as the 'Phillips curve.'

(7) Goodwin also assumes that "all wages [are] consumed, [and] all profits are saved and invested." (Goodwin, pg. 54) Grasselli and Lima show that this means that the change in capital is equal to output, minus wages paid to labour and depreciation of capital (Say's law).

$$\frac{dK}{dt}= Y - wL - \delta K $$

(8) Grasselli and Lima define $\omega$, the share of the economy paid to labour, as follows:

$$\omega(t) = \frac{w(t)L(t)}{a(t)L(t)} = \frac{w(t)}{a(t)} $$

This means the share of economic output paid to labour through wages is the real wage rate divided by labour productivity.

Given these assumptions, it is now possible to define the growth rates of the economy ($Y$), the wage/labour share of the economy ($\omega$), and the employment rate ($\lambda$).

#### Derivations:

(a) Growth rate of the economy

By definition, the growth rate of the economy is $\frac{dY/dt}{Y}$

$$\frac{dY/dt}{Y} = \frac{d(K/v)/dt}{Y} $$

by assumption (5).

$$ = \frac{dK/dt}{Yv} $$
$$ = \frac{Y-wL-\delta K}{Yv} $$
$$ = \frac{(1-\omega)Y-\delta K}{Yv} $$
because by assumption (8), $\omega(t) = \frac{w(t)L(t)}{Y}$ which implies $w(t)L(t)=\omega(t)Y$.
$$ = \frac{1 -\omega}{v} - \frac{\delta K}{Yv} $$

By assumption (5), $K(t)/v=Y$, so

$$ = \frac{1 -\omega}{v} - \frac{\delta Y}{Y} $$
$$ = \frac{1 -\omega}{v} - \delta $$

Therefore the growth rate of the economy is 

$$\frac{dY/dt}{Y} = \frac{1 -\omega}{v} - \delta $$

as in Grasselli and Lima (2012).

(b) Growth rate of share of the economy paid to labour (wage share)

Growth rate of the wage share is $\frac{d\omega/dt}{\omega}$

$$\frac{d\omega/dt}{\omega} = \frac{d(w/a)/dt}{w/a} $$

by assumption (8).

$$ = \frac{dw/dt}{w} - \frac{da/dt}{a} $$

$$ = \frac{w(f(\lambda))}{w} - \frac{\alpha a}{a} $$

by assumptions (6) and (4).


Therefore the growth rate of the wage share is 

$$\frac{d\omega/dt}{\omega} = f(\lambda) - \alpha$$

as in Grasselli and Lima (2012).

(c) Growth rate of the employment rate

By definition, the growth rate of the employment rate is $\frac{d\lambda/dt}{\lambda}$

$$\frac{d\lambda/dt}{\lambda} = \frac{d(L/N)/dt}{L/N} $$

by assumption (3).

$$ = \frac{dL/dt}{L} - \frac{dN/dt}{N} $$

By assumption (5), $L(t)=\frac{Y(t)}{a(t)}$, so

$$ = \frac{d(Y/a)/dt}{Y/a} - \frac{dN/dt}{N} $$

$$ = \frac{dY/dt}{Y} - \frac{da/dt}{a} - \frac{dN/dt}{N} $$

Here, the first term is the growth rate of the economy (derived above), and the second terms are the growth rates of labour productivity and total labour force, given by assumptions (4) and (2), respectively.


Therefore the growth rate of employment rate is 

$$\frac{d\lambda/dt}{\lambda} = \frac{1 -\omega}{v} - \delta - \alpha - \beta $$

as in Grasselli and Lima (2012).

If we rearrange the growth rates of wage share ($\omega$) and employment rate ($\lambda$), we get the Goodwin model as in Grasselli and Lima (2012):

$$\frac{d\omega/dt}{\omega} = f(\lambda) - \alpha$$ and $$\frac{d\lambda/dt}{\lambda} = \frac{1 -\omega}{v} - \delta - \alpha - \beta $$

imply

$$\frac{d\omega}{dt} = \omega(f(\lambda) - \alpha)$$ and $$\frac{d\lambda}{dt} = \lambda \left( \frac{1 -\omega}{v} - \delta - \alpha - \beta \right) $$


### Parameter estimation:

We wish to estimate the following parameters for the Goodwin model:

$\lambda$ = productivity growth rate (annual) 

$\beta$ = the population growth rate (annual)

$\delta$ = the constant depreciation rate (annual) 

$v$ = capital to output ratio (annual)

It is important that our parameters follow the condition: $\frac{1}{v} - \alpha - \beta - \delta > 0 $. The reason for this is explained below when we examine the stability of the Goodwin model, but this condition ensures that the nonzero equilibrium, and not the zero equilibrium, is stable. This condition means the marginal product of capital ($\frac{1}{v}$) must be greater than sum of productivity growth rate, the population growth rate, and the constant depreciation rate. 

First, we estimate $\beta$. We examined data on Canadian population growth from 2005 to 2016 (Statistics Canada, 2017, "Canada at a Glance 2017: Population."). We find the average population growth rate over this period.

The mean population growth rate is 0.010489166666666667, which we round to 0.0105 (1.05%).

Second, we compute the captial-output ratio, $v$. Incremental capital-output ratio (ICOR) is a ratio of investment to income (GDP) growth. Investment is the growth rate of capital, so the ICOR is the capital to income ratio, $v$, that we seek. Usually, most countries' ICORs are in the neighbourhood of 3, so we expect our estimate to be in that neighbourhood. Also, both Keen (1995) and Grasselli and Lima (2012) use $v=3$. We get data on Canadian ICOR from the OECD and find the mean value from 2005 to 2011 (Kirby, 2015).

```python
YR = [2005.0, 2006.0, 2007.0, 2008.0, 2009.0, 2010.0, 2011.0] 
ICOR = [243.732922, 249.505612, 256.665461, 267.445692, 284.592068, 286.841594, 292.048565]

plt.plot(YR,ICOR)
plt.show()
mean = sum(ICOR)/len(YR)
print(mean/100)
```

The mean capital to output ratio is 2.686902734285715, which we round to $v=2.69$.

It is more difficult to estimate $\alpha$ and $\delta$, because $\alpha$, the technology growth rate, is unobserved, and all estimates of Canadian depreciation rates are much higher than those used in Keen (1995) and Grasselli and Lima (2012). We choose $\alpha = 0.02$, as this is the average of the parameters used in Keen (1995) and Grasselli and Lima (2012). We choose a depreciation rate of $\delta = 0.02$, as this was the higher of the two estimates in the literature.

Lastly, we confirm that our choice of parameters satisfies the condition $\frac{1}{v} - \alpha - \beta - \delta > 0 $:

$$\frac{1}{v} - \alpha - \beta - \delta $$
$$=\frac{1}{2.69} - 0.02 - 0.0105 - 0.02$$
$$= 0.32124721189 > 0 $$

So we are assured that our chosen parameters are reasonable.

In summary, the estimated parameters are:

$\alpha = 0.02$

$\beta = 0.0105$

$\delta = 0.02$

$v=2.69$

### Analysis

#### Equilibria:

Set both equations equal to zero:

$$ 0 = \omega(f(\lambda)-\alpha) $$
$$ 0 =\lambda \left(\frac{1-\omega}{v}-\delta-\alpha-\beta \right) $$

From the first equation, either (i) $\omega=0$ or (ii) $f(\lambda)-\alpha=0$. Solving (ii) for $\lambda$ give $\lambda = f^{-1}(\alpha)$. 

Therefore the first equation gives two nullclines, $\omega=0$ and $\lambda = f^{-1}(\alpha)$.

From the second equation, either (i) $\lambda=0$ or (ii) $\frac{1-\omega}{v}-\delta-\alpha-\beta = 0$. From (ii),
$$\frac{1-\omega}{v} =\delta + \alpha + \beta$$
$$1-\omega=v(\delta + \alpha + \beta)$$
$$\omega=1-v(\delta + \alpha + \beta)$$

Therefore the second equation gives two nullclines, $\lambda=0$ and $\omega = 1-v(\delta + \alpha + \beta)$.

There are two equilibrium points, $(\omega^{*}, \lambda^{*}) = (0,0)$ and $(1-v(\delta + \alpha + \beta), f^{-1}(\alpha))$.

Given the parameters we have chosen for this project, the equilibrium points are $(\omega^{*}, \lambda^{*}) = (0,0)$ and $(0.86, 0.95)$.

#### Stability:

The Jacobian matrix for this system is
$$
\begin{split}
J(\omega,\lambda) = \left( \begin{array}{cc}
f(\lambda) - \alpha & \omega f'(\lambda) \\
\frac{-\lambda}{v} & \frac{1-\omega}{v}-\delta-\alpha-\beta \\
\end{array}\right)
\end{split}
$$

First we consider the trivial equilibrium. We evaluate the Jacobian matrix at $(\omega^{*}, \lambda^{*}) = (0,0)$, and find
$$
\begin{split}
J(0,0) = \left( \begin{array}{cc}
f(0) - \alpha & 0 \\
0 & \frac{1}{v} -\alpha - \beta -\delta \\
\end{array}\right)
\end{split}
$$

This is a diagonal matrix, so we can read the eigenvalues off of the diagonal. The eigenvalues are $f(0) - \alpha$ and 
$1/v -\alpha - \beta -\delta$.

If we assume that $f(0) < \alpha$ and $1/v -\alpha - \beta -\delta > 0$, then the first eigenvalue is negative while the second one is positive, so (0,0) is unstable (locally a saddle point).

Why do we make these assumptions?

It is reasonable to assume that $f(0) < \alpha$, because $f(\lambda)$ is the Phillips curve and should have a positive x-intercept to have a reasonable economic interpretation (the x-intercept is the level of employment at which workers accept a constant wage rate, so must be between 0 and 1). This means $f(0) < 0 <\alpha$ because $\alpha$ is a positive constant.

The second assumption is that $\frac{1}{v} -\alpha - \beta -\delta > 0$. This means $\frac{1}{v} > \alpha + \beta + \delta$.

This assumption says that the ratio of capital to output is greater than the sum of the productivity growth rate, the population growth rate, and the capital depreciation rate. The productivity growth rate, the population growth rate, and the capital depreciation rate are all small in practice, each under 5%, while the capital-to-output ratio should be greater than 1 because capital is assumed to be productive (capital should create significantly more income than its value). In the literature, the capital-to-output ratio is assumed to be 3, so $\frac{1}{v} > \alpha + \beta + \delta$ for any reasonable choices of the productivity growth rate ($\alpha$), the population growth rate ($\beta$), and the capital depreciation rate ($\delta$).

Also, it is worth noting that if we had $\frac{1}{v} -\alpha - \beta -\delta < 0$, then the zero equilibrium would be stable, which is problematic for an economic model.


At the nonzero equilibrium, $(\omega^{*}, \lambda^{*}) = (1-v(\delta + \alpha + \beta), f^{-1}(\alpha))$, the Jacobian is:
$$
\begin{split}
J(1-v(\delta + \alpha + \beta), f^{-1}(\alpha)) = \left( \begin{array}{cc}
0 & (1-v(\delta + \alpha + \beta) f'(f^{-1}(\alpha)) \\
\frac{-f^{-1}(\alpha)}{v} & 0\\
\end{array}\right)
\end{split}
$$

The characteristic equation is $$0=\lambda^{2}+\frac{f^{-1}(\alpha)(1-v(\delta + \alpha + \beta) f'(f^{-1}(\alpha))}{v}$$

So

$$\lambda^{2}=-\frac{f^{-1}(\alpha)(1-v(\delta + \alpha + \beta) f'(f^{-1}(\alpha))}{v}$$

$$\lambda^{2}=\sqrt{-\frac{f^{-1}(\alpha)(1-v(\delta + \alpha + \beta) f'(f^{-1}(\alpha))}{v}}$$

so if $1>v(\delta + \alpha + \beta)$ (same assumption as for the (0,0) equilibrium), the eigenvalues are pure imaginary. This means the equilibrium point is non-hyperbolic. The Jacobian cannot tell us if the equilibrium is stable or not (the linearization fails if the eigenvalues have zero real part), but numerically we see that it has periodic orbits around it.

### Numerical Analysis

#### Code for the model:

To conduct any numerical analysis of the Goodwin model, we need a Phillips curve, $f(\lambda)$. $f(\lambda)$ measures wage change. Its economic meaning is that workers accept constant wages at the x-intercept, decreasing wages below that employment level, and demand increasing wages above that level.

Grasselli and Lima (2012) give conditions that the Phillips curve must follow. The Phillips curve can be any continuously differentiable function on the interval (0,1) provided that for $0\leq \lambda(t) \leq 1 $

$$ f'(\lambda)>0 $$
$$ f(0)<\alpha $$
$$ \lim_{\lambda\to1^{-}} f(\lambda) = \infty $$

We reviewed Canadian monthly unemployment rates (seasonally adjusted) since the year 2000 (Statistics Canada, 2017. "Table 282-0087"). The unemployment rate has varied between a low of 5.8 in October 2007 and a high of 8.7 from June to August 2009. Given this information, we select an unemployment rate of 7 percent as the rate at which workers accept a constant wage, and assume they demand higher wages above this unemployment rate and accept wage cuts if unemployment is higher. 

(It is interesting to note that even though the model limits $\lambda$ to $0 \leq \lambda \leq 1$, in recent times (a time that includes both good economic times and a bad recession) the employment rate has in fact been $91.3 \leq \lambda \leq 94.2$, so realistically $\lambda$ is much less variable than assumed.)

We also need a functional form. We could use linear approximation through (0.93,0) but then we do not have asymptotic behaviour at 100 percent employment as is desired. 

The following function is in the form Keen (1995) used, and fits the 3 criteria given by Grasselli and Lima, while also fitting our criteria that workers accept constant wages at 7 percent unemployment:

$$ f(\lambda) = \frac{0.0001}{(1-\lambda)^2} - 0.0204 $$

The Phillips curve we chose looks like this:

```python
# defines and graphs the Phillips curve for illustrative purposes
tvec = np.arange(0,1,0.01)

def f(x,t):
    return 0.0001/((1-x)**2)-0.0204
xvec = np.arange(0,1,0.01)

f_lambda = f(xvec,0)

%matplotlib inline
plt.plot(xvec,f_lambda)
plt.axhline(0,c="black")
plt.axvline(1,c="black",linestyle="--")
plt.title('Phillips Curve')
plt.xlabel('lambda')
plt.ylabel('f')
plt.ylim((-0.05,0.2));
```

This fits the conditions of Grasselli and Lima (2012).

We can now write the code for the Goodwin model:

```python
# defines Goodwin model with 4 parameters and 2 variables
def goodwin(x, t, params):
    a, b, d, v  = params
    W, L = x
    return(np.array([W*(0.0001/((1-L)**2)-0.0204-a),
                    L*((1-W)/v-a-b-d)]))
```

#### Time-dependent solutions:

We use the parameters estimated above. For the initial employment rate, we choose $\lambda=0.937$, since Canada's unemployment rate is currently 6.3 percent (Statistics Canada, 2017, "Table 282-0087"). We choose $\omega=0.96$ as in Grasselli and Lima (2012), as the wage share is harder to estimate from data.

```python
tvec = np.arange(0,50,0.1)
params1 = (0.02,0.0105,0.02,2.69)
attempt_1_2d = scipy.integrate.odeint(
    goodwin,
    y0 = (0.96,0.937),
    t = tvec,
    args = (params1,))
```

```python
# what attempt 1 looks like
%matplotlib inline
plt.plot(attempt_1_2d);
plt.legend(["$\omega$","$\lambda$"])
plt.xlabel("time")
plt.ylabel("$\lambda, \omega$");
```

We note that employment leads wage share, which makes sense because from the Phillips curve when employment is high, workers demand higher wages, leading to a higher wage share.

```python
# Change the initial conditions and plot the time dependent solution
attempt_2_2d = scipy.integrate.odeint(
    goodwin,
    y0 = (0.9,0.9),
    t = tvec,
    args = (params1,))
%matplotlib inline
plt.plot(attempt_2_2d);
plt.legend(["$\omega$","$\lambda$"])
plt.xlabel("time")
plt.ylabel("$\lambda, \omega$");
```

#### Phase plane:

We plot the phase plane for four solutions using different initial conditions:

```python
# third set of IC
attempt_3_2d = scipy.integrate.odeint(
    goodwin,
    y0 = (0.85,0.7),
    t = tvec,
    args = (params1,))

# fourth set of IC
attempt_4_2d = scipy.integrate.odeint(
    goodwin,
    y0 = (0.75,0.85),
    t = tvec,
    args = (params1,))

# plot the phase plane
%matplotlib inline
plt.axes().set_aspect('equal', 'datalim')
plt.plot(attempt_1_2d[:,0],attempt_1_2d[:,1]);
plt.plot(attempt_2_2d[:,0],attempt_2_2d[:,1],c="red");
plt.plot(attempt_3_2d[:,0],attempt_3_2d[:,1],c="green");
plt.plot(attempt_4_2d[:,0],attempt_4_2d[:,1],c="orange");
plt.xlabel("$\omega$")
plt.ylabel("$\lambda$");
```

The phase diagram shows that the solutions are closed orbits around the equilibrium point $(\omega^{*}, \lambda^{*}) = (1-v(\delta + \alpha + \beta), f^{-1}(\alpha)) = (0.86, 0.95)$. This indicates that the nonzero equilibrium is stable (but not asymptotically stable).

We can also show the equilibrium using nullclines. To do that, we make a change to the model code to eliminate the possibility of dividing by zero.

```python
def goodwina(x, t, params):
    a, b, d, v  = params
    W, L = x
    if L == 1:
        dW = 1
    else:
        dW=W*(0.0001/((1-L)**2)-0.0204-a)
    return(np.array([dW,
                    L*((1-W)/v-a-b-d)]))
## eliminates the possibility of dividing by zero in the Goodwin model
```

```python
npts = 51
xvec = np.linspace(-0.1,1.01,npts)
yvec = np.linspace(-0.1,1.01,npts)
ns = 2 
pp = np.zeros((npts,npts,ns))
for i in range(npts):
    for j in range(npts):
        pp[i,j,:] = goodwina((xvec[i],yvec[j]),0,params1)
DW = pp[:,:,0].T
DL = pp[:,:,1].T
```

```python
%matplotlib inline
## contour plot with contour lines at zero only
plt.contour(xvec,yvec,DW,levels=[0],colors=["red"])
# omega nullclines are red
plt.contour(xvec,yvec,DL,levels=[0],colors=["blue"])
# lambda nullclines are blue
plt.ylim(-0.1,1.01)
plt.xlim(-0.1,1.01)
plt.xlabel("$\omega$")
plt.ylabel("$\lambda$");
```

The equilibrium points occur at the intersection of the red and blue lines. They are $(\omega^{*}, \lambda^{*}) = (0,0)$ and $(0.86, 0.95)$, as expected.

```python
# plot the phase plane with both nullclines and orbits (orbits are all in green)
%matplotlib inline
plt.contour(xvec,yvec,DW,levels=[0],colors=["red"])
plt.contour(xvec,yvec,DL,levels=[0],colors=["blue"])
plt.ylim(-0.1,1.05)
plt.xlim(-0.1,1.1)
plt.plot(attempt_1_2d[:,0],attempt_1_2d[:,1],c="green");
plt.plot(attempt_2_2d[:,0],attempt_2_2d[:,1],c="green");
plt.plot(attempt_3_2d[:,0],attempt_3_2d[:,1],c="green");
plt.plot(attempt_4_2d[:,0],attempt_4_2d[:,1],c="green");
plt.xlabel("$\omega$")
plt.ylabel("$\lambda$");
```

This shows clearly that the closed orbits encircle the nonzero equilibrium (0.86, 0.95).

#### Model with income

Since we are modelling an economy, we wish to see the path of the economy over time. The path of the economy is the path of income (Y), which is generally measured by GDP. In real life, Y grows over time, with small variations from the trend. These variations are booms and busts (recessions).
We add Y to the model. The growth rate of Y was found in the model derivation.

```python
# adds income (Y) to the 2D model
def goodwiny(x, t, params):
    a, b, d, v  = params
    W, L, Y = x
    return(np.array([W*(0.0001/((1-L)**2)-0.0204-a),
                    L*((1-W)/v-a-b-d),
                    Y*((1-W)/v-d)]))
```

```python
# here we repeat attempt 1 (our initial conditions) with initial income of 100
tvec = np.arange(0,50,0.1)
attempt_y_2d = scipy.integrate.odeint(
    goodwiny,
    y0 = (0.96,0.937,100),
    t = tvec,
    args = (params1,))
```

```python
# to plot the time path of wage share, employment, and income, we use two different scales
# We use mpatches to create a legend

fig, ax1 = plt.subplots()
ax1.plot(tvec, attempt_y_2d[:,0], 'steelblue')
ax1.plot(tvec, attempt_y_2d[:,1], 'darkorange')
ax1.set_xlabel('time')
ax1.set_ylabel('$\lambda$, $\omega$')

ax2 = ax1.twinx()
ax2.plot(tvec, attempt_y_2d[:,2], 'indianred')
ax2.set_ylabel('Y')

blue_patch = mpatches.Patch(color='steelblue', label='$\omega$')
orange_patch = mpatches.Patch(color='darkorange', label='$\lambda$')
red_patch = mpatches.Patch(color='indianred', label='Y')
plt.legend(handles=[blue_patch,orange_patch,red_patch],loc=4)

fig.tight_layout()
plt.show()
```

Y grows over time with small deviations from trend, as we would expect.

## Keen Model (3D)

### Model Construction

#### Model:

The second model is the 'Keen', a 3-dimensional model first elaborated by economist Steve Keen. The model is:

$$\frac{d\omega}{dt} = \omega (f(\lambda) - \alpha) $$
$$\frac{d\lambda}{dt} = \lambda \left( \frac{\kappa(1 -\omega - r\eta)}{v} - \delta - \alpha - \beta \right)$$
$$ \frac{d\eta}{dt} = \eta \left( r - \frac{\kappa(1-\omega-r\eta)}{v} + \delta \right) + \kappa(1-\omega-r\eta)- (1-\omega) $$

#### New parameters:

Let $r$ be the real interest rate (units are /year, assumed constant).

Let $D$ be the total amount of debt in the economy (unit is dollars).

Let $\pi$ be net profit share ($\pi=1-\omega-r\eta$) (unitless).

Let $\kappa(\pi)$ be investment. It is an increasing function, and is unitless.

####  Variables:

Let $\lambda$ be the employment rate, the number of employed persons divided by the total labour force ($\lambda=\frac{L}{N}$) (unitless, need $0 \leq \lambda \leq 1 $ for this to make sense).

Let $\omega$ be the share of the economy paid to labour through wages (unitless, $0 \leq \omega$ for Keen model).

Let $\eta$ be the debt ratio in the economy ($\eta=\frac{D}{Y}$) (has units of years, because $D$ has units of dollars and $Y$ has units of dollars/year, $0 \leq \eta$).

### Model derivation:

#### Assumptions:

We introduce the following three additional economic assumptions, as in Grasselli and Lima (2012):

(9) Net profit is $\pi = 1-\omega - r\eta $ (what is left after wages and interest are paid).

(10) In the Goodwin model, one of the key assumptions is that wages are determined by the Phillips curve - the assumption that workers demand higher real wages when the employment rate is higher. Keen (1995) introduces a similar assumption for the behaviour of capitalists (investors). He assumes that investors are more likely to invest when economic conditions are good (booms) rather when economic conditions are poor (busts) (Keen, 1995, pg. 614). Therefore Keen models investment as an increasing nonlinear function of profit, $\kappa(\pi)$.

(11) Keen assumes the change in capital stock to be new investment minus depreciation:

$$ \frac{dK}{dt}=\kappa (\pi) Y - \delta K =\kappa (1-\omega - r\eta) Y - \delta K $$


#### Derivation:

(a) growth rate of the economy

Recall that by assumption (5), $ \frac{K(t)}{v}=Y(t) $. So the growth rate of the economy is:

$$\frac{dY/dt}{Y} = \frac{d(K(t)/v)/dt}{Y} $$
$$ = \frac{d(K(t))/dt}{Yv} $$
$$ = \frac{Y-wL-\delta K}{Yv} $$
$$ = \frac{(\kappa(\pi)Y-\delta K}{Yv} $$
$$ = \frac{\kappa(\pi)}{v}-\frac{\delta K}{Yv} $$
$$ = \frac{\kappa(\pi)}{v} - \delta $$
$$ = \frac{\kappa(1-\omega-r\eta)}{v} - \delta $$

as in Grasselli and Lima (2012).

(b) growth rate of the wage share

The growth rate of the wage share is unchanged by the inclusion of debt in the model, so it remains

$$\frac{d\omega/dt}{\omega} = f(\lambda) - \alpha$$

(c) growth rate of the employment rate

The growth rate of the employment rate is $\frac{d\lambda/dt}{\lambda}$. Recall from above that,

$$\frac{d\lambda/dt}{\lambda} = \frac{dY/dt}{Y} - \frac{da/dt}{a} - \frac{dN/dt}{N} $$

So, using Keen's growth rate of the economy, the growth rate of the employment rate is

$$\frac{d\lambda/dt}{\lambda} = \frac{\kappa(\pi)}{v} - \delta - \alpha - \beta $$

$$ = \frac{\kappa(1-\omega-r\eta)}{v} - \delta - \alpha - \beta $$

as in Grasselli and Lima (2012).

(d) growth rate of the debt share

We also need an equation for growth rate of the debt ratio (the new variable in the Keen model).

How does total debt change over time? It is the difference between new investment and profits, so 

$$ \frac{dD}{dt} = \kappa (\pi)Y - \pi Y $$

Therefore the growth rate of the debt ratio is

$$\frac{d\eta/dt}{\eta} = \frac{d(D/Y)/dt}{(D/Y)} $$

$$ = \frac{dD/dt}{D} - \frac{dY/dt}{Y} $$

$$ = \frac{\kappa(\pi)Y - \pi Y}{D} - (\frac{\kappa(\pi)}{v} - \delta) $$

$$ = \frac{\kappa(\pi)- \pi}{\eta} - \frac{\kappa(\pi)}{v} + \delta $$

$$ = \frac{\kappa(1-\omega-r\eta)- (1-\omega-r\eta)}{\eta} - \frac{\kappa(1-\omega-r\eta)}{v} + \delta $$

as in Grasselli and Lima (2012).

To get the differential equation for the debt ratio, we arrange the growth rate as follows:

$$\frac{d\eta/dt}{\eta} = \frac{\kappa(1-\omega-r\eta)- (1-\omega-r\eta)}{\eta} - \frac{\kappa(1-\omega-r\eta)}{v} + \delta $$

==>

$$\frac{d\eta}{dt} = \eta [\frac{\kappa(1-\omega-r\eta)- (1-\omega-r\eta)}{\eta} - \frac{\kappa(1-\omega-r\eta)}{v} + \delta] $$

$$ = \eta [ - \frac{\kappa(1-\omega-r\eta)}{v} + \delta] + \kappa(1-\omega-r\eta)- (1-\omega-r\eta) $$

$$ = \eta [r - \frac{\kappa(1-\omega-r\eta)}{v} + \delta] + \kappa(1-\omega-r\eta)- (1-\omega) $$

which is the differential equation used in Grasselli and Lima (2012).

Therefore, we have the following system of differential equations:

$$\frac{d\omega}{dt} = \omega (f(\lambda) - \alpha) $$
$$\frac{d\lambda}{dt} = \lambda \left( \frac{\kappa(1 -\omega - r\eta)}{v} - \delta - \alpha - \beta \right)$$
$$ \frac{d\eta}{dt} = \eta \left( r - \frac{\kappa(1-\omega-r\eta)}{v} + \delta \right) + \kappa(1-\omega-r\eta)- (1-\omega) $$

#### Parameter estimation:

The only additional parameter in this model is $r$, the real interest rate. The real interest rate is the rate of interest a saver/lender receives after allowing for inflation. The real interest rate is approximately the nominal interest rate minus the inflation rate. We estimate the Canadian real interest rate using data from 2005 to 2016 (The World Bank, 2017). We compute the mean real interest rate over this time interval.

```python
YR = [2005.0, 2006.0, 2007.0, 2008.0, 2009.0, 2010.0, 2011.0, 2012.0, 2013.0, 2014.0, 2015.0, 2016.0]
INTRST = [1.231,3.118,2.696,0.696,4.8,-0.262,-0.234,1.757,1.393,1.083,3.582,2.082]

plt.plot(YR,INTRST)
plt.show()
mean = sum(INTRST)/len(YR)
print(mean/100)
```

The mean value is 0.01828, which we round to 0.02. Therefore our estimate of the real interest rate is 2%.  

In summary, we use the following parameters in the Keen model

$\alpha = 0.02$, $\beta = 0.0105$, $\delta = 0.02$, $v=2.69$, $r=0.02$.

### Analysis

#### Equilibria:

Set the equations equal to zero:

$$ 0 = \omega(f(\lambda)-\alpha) $$
$$ 0 =\lambda \left( \frac{\kappa(1 -\omega - r\eta)}{v} - \delta - \alpha - \beta \right) $$
$$ 0 =\eta \left( r-\frac{\kappa(1-\omega-r\eta)}{v}+\delta \right) +\kappa(1-\omega-r\eta)-(1-\omega)$$

From the first equation, either (i) $\omega=0$ or (ii) $f(\lambda)-\alpha=0$. Solving (ii) for $\lambda$ give $\lambda = f^{-1}(\alpha)$. 

Therefore the first equation gives two nullclines, $\omega=0$ and $\lambda = f^{-1}(\alpha)$.

From the second equation, either (i) $\lambda = 0$ or (ii) $0 = \frac{\kappa(1 -\omega - r\eta)}{v} - \delta - \alpha - \beta$.

From (ii), $$\frac{\kappa(1 -\omega - r\eta)}{v} = \delta + \alpha + \beta$$

 $$ \kappa(1 -\omega - r\eta) = v(\delta + \alpha + \beta)$$
 
 $$ 1 -\omega - r\eta = \kappa^{-1}(v(\delta + \alpha + \beta))$$
 
So either
$$\omega = 1 - \kappa^{-1}(v(\delta + \alpha + \beta)) - r\eta$$ or $$ \eta = \frac{1 - \kappa^{-1}(v(\delta + \alpha + \beta)) - \omega}{r}$$

Therefore the second equation also gives two nullclines, $\lambda = 0$ and $\omega = 1 - \kappa^{-1}(v(\delta + \alpha + \beta)) - r\eta$.

From the third equation, 

$$ 0 =\eta \left( r-\frac{\kappa(1-\omega-r\eta)}{v}+\delta \right) +\kappa(1-\omega-r\eta)-(1-\omega)$$

$$ \eta \left( r-\frac{\kappa(1-\omega-r\eta)}{v}+\delta \right) = 1-\omega - \kappa(1-\omega-r\eta)$$

This could be solved for either $\omega$ or $\eta$. Solving for $\eta$:

$$ \bar{\eta} = \frac{1-\omega - \kappa(1-\omega-r\eta)}{\left( r-\frac{\kappa(1-\omega-r\eta)}{v}+\delta \right)}$$

Two potential equilibrium points are $(\omega^{*},\lambda^{*},\eta^{*}) = (0,0,\bar{\eta})$ and $(\omega^{*},\lambda^{*},\eta^{*}) = ((1 - \kappa^{-1}(v(\delta + \alpha + \beta)) - r\eta),f^{-1}(\alpha),\bar{\eta})$

The first equilibrium is trivial. The second can be simplified to $(\omega^{*},\lambda^{*},\eta^{*}) = (1 - \kappa^{-1}(v(\delta + \alpha + \beta)) - r\frac{v(\delta + \alpha + \beta) - \kappa^{-1}(v(\delta + \alpha + \beta))}{\alpha + \beta},f^{-1}(\alpha),\frac{v(\delta + \alpha + \beta) - \kappa^{-1}(v(\delta + \alpha + \beta))}{\alpha + \beta})$.

There is a third potential equilibrium.

Assuming that $\omega = 0$ and $\lambda \neq 0$, then

$$
\begin{split}
0 & = \lambda \left( \frac{\kappa(1-r\eta)}{\nu} - \alpha -\beta - \delta \right) \\
\frac{\kappa(1-r\eta)}{v} & = \alpha + \beta + \delta \\
1-r\eta & = \kappa^{-1}(v(\alpha + \beta + \delta)) \\
\eta & = \frac{\kappa^{-1}(v(\alpha + \beta + \delta)) + 1}{r}
\end{split}
$$

But $\eta$ must also be of the form of $\eta_0$ in order for $\eta$ to be at equilibrium. Substituting for $\eta$ in $(*)$, we find that this equilibrium exists only when the paramaters satisfy the following equation.

$$
1- r\frac{\nu(\alpha + \beta +\delta) - \kappa^{-1}(v(\alpha + \beta +\delta))}{\alpha + \beta} = \kappa^{-1}(v(\alpha + \beta + \delta))
$$

Since this equilibrum only exists when this specific condition is met for the parameters, we consider this equilbrium to be not meaningful.

#### Stability:

The Jacobian for the Keen model is given by:
$$
\begin{split}
J(\omega^*,\lambda^*, \eta^*)  = \left( \begin{array}{ccc}
 f(\lambda) -\alpha & \omega f' (\lambda) &0 \\
 -\frac{\lambda \kappa'(1 - \omega -r\eta)}{v} & \frac{\kappa(1 - \omega -r\eta)}{v} -(\alpha + \beta + \delta) & -\frac{r \lambda \kappa'(1 - \omega - r\eta)}{v} \\
\frac{(d - v) \kappa'(1 - \omega -r\eta) + v}{v} & 0 & \frac{v(r + \delta) - \kappa(1 - \omega - r\eta) + r(d-v)\kappa'(1 - \omega - r\eta)}{v} \\ 
\end{array}\right)
\end{split}
$$

At the equilibrium point $(\omega,\lambda,\eta) = (0,0,\eta_0)$, the Jacobian is given by

$$
\begin{split}
J(0, 0, \eta_0)  = \left( \begin{array}{ccc}
 f(0) -\alpha & 0 & 0 \\
 0 & \frac{1}{v}\left(\kappa(1 -r\eta_0) -v(\alpha + \beta + \delta) \right)& 0 \\
\frac{(d - v) \kappa'(1 -r\eta_0) + v}{v} & 0 & \frac{v(r + \delta) - \kappa(1 - r\eta_0) + r(\eta_0-v)\kappa'(1 - r\eta_0)}{v} \\ 
\end{array}\right)
\end{split}
$$

Since this matrix is triangular, its diagonal gives its eigenvalues. The first eigenvalue is negative, if we continue the assuption that $f(0) < \alpha$. Then if either of the remaining eigenvalues is positive, this equilibrium is unstable. We expect this equilibrium to be unstable for two reasons. Firstly, even if debt remains stable, this equilibrium now resembles the trivial equilibrium of the Goodwin model, which is unstable.

The second reason we can expect this equilibrium to be unstable is that we expect $\kappa(1-r\eta_0)$ to have a similar value to $v(\alpha + \beta + \delta)$. If the second eigenvalue is negative, then $\eta_0$ must be large, so that $\kappa(1 - r\eta_0) < v(\alpha + \beta + \delta)$. But then we expect 

$$\frac{v(r + \delta) - \kappa(1 - r\eta_0) + r(\eta_0-v)\kappa'(1 - r\eta_0)}{v}$$

to be positive, because $\kappa$ is increasing, $v$ is relatively small, and $-\kappa(1-r\eta_0)$ has a relatively small absolute value, since $v, \alpha, \beta, \delta$ are all relatively small quantities.

Similarly, if the third eigenvalue is negative, we expect $\eta_0$ to be small, and then $\kappa(1 - r\eta_0) > v(\alpha + \beta + \delta)$. Therefore, we can expect this equilibrium to be unstable.

We numerically explore the stability of the equilibrium of the form $\omega^* = 1 - \kappa^{-1}(v(\alpha + \beta + \delta)) -r\eta$, $\lambda^* = f^{-1}(\alpha)$, and $\eta^* = \frac{v(\alpha + \beta + \delta) -\kappa^{-1}(v(\alpha + \beta + \delta))}{\alpha + \beta}$.

### Numerical Analysis

To conduct any numerical analysis of the Keen model, we need to define an investment function, $\kappa(\pi)$, where $\pi=1-\omega-r\eta$. 

Grasselli and Lima (2012) also give conditions that this investment function must follow.  $\kappa(\pi)$ must be a continuously differentiable function satifying the following 3 conditions:

$$ \kappa'(\pi)>0 $$
$$ \lim_{\pi\to-\infty} \kappa(\pi) < v(\alpha + \beta + \delta) < \lim_{\pi\to+\infty} \kappa(\pi) $$
$$ \lim_{\pi\to-\infty} \pi^2 \kappa(\pi) = 0 $$


We could choose any reasonable exponential function (it shouldn't blow up on (0,1)). However, the model is very sensitive to the choice of investment function. For example, $\kappa(\pi) = e^{20\pi-3}$, $\kappa(\pi) = e^{20\pi-4}$, and $\kappa(\pi) = e^{20\pi-5}$ cause the model to have very different dynamics (in the first case, debt share explodes; in the second, debt share levels off a little above 1; in the third, it falls to below zero, levelling off at about -0.4). It would be interesting to further investigate these differences, but we unfortunately do not have time to pursue this. We choose $\kappa(\pi) = e^{20\pi-4.5}$ which is very similar to that chosen by Grasselli and Lima (2012) (they used $\kappa(\pi) = e^{20\pi-5}$). We can't use the same investment function as Grasselli and Lima, because it gives us a negative equilibrium debt level.

```python
# plots the investment function, kappa, for illustrative purposes
def kappa(x,t):
    return(np.exp(20*(x)-4.5))

xvec = np.arange(0,1,0.01)

a = kappa(xvec,0)

%matplotlib inline
plt.plot(xvec, a);
```

Using this investment function and the Phillips curve we used before, we can code the model. We use the parameters we estimated.

```python
def keen(x, t, params):
    a, b, d, v, r  = params
    W, L, D = x
    return(np.array([W*(0.0001/((1-L)**2)-0.0204-a),
                    L*((np.exp(20*(1-W-r*D)-4.5)/v)-a-b-d),
                    D*(r-(np.exp(20*(1-W-r*D)-4.5)/v)+d)+np.exp(20*(1-W-r*D)-4.5)-(1-W)]))
```

## Financial Speculation Model (4D)

### Model Construction

#### Model:

Grasselli and Lima (2012) extend Keen's model to allow for financial speculation. The 4-dimensional 'Financial Speculation' model is:

$$\frac{d\omega}{dt} = \omega [f(\lambda) - \alpha] $$

$$\frac{d\lambda}{dt} = \lambda \left( \frac{\kappa(1 -\omega - r\eta)}{v} - \delta - \alpha - \beta \right) $$

$$ \frac{d\eta}{dt} = \eta \left( r - \frac{\kappa(1-\omega-r\eta)}{v} + \delta \right) + \kappa(1-\omega-r\eta)- (1-\omega) + p$$

$$ \frac{dp}{dt} = p \left( \psi(\frac{\kappa(1-\omega-r\eta)}{v} - \delta) - \frac{\kappa(1-\omega-r\eta)}{v} + \delta \right) $$

#### Parameters:

Let $P$ be total financial speculation in the economy (measured in dollars).

There is also a new function:

$\psi$: an increasing, non-negative function of the growth rate of the economy, $\frac{\kappa(1-\omega-r\eta)}{v} - \delta$ (units of /year)

#### Variables:

Let $\lambda$ be the employment rate, the number of employed persons divided by the total labour force ($\lambda=L/N$). (unitless, need $0 \leq \lambda \leq 1 $).

Let $\omega$ be the share of the economy paid to labour through wages (unitless, $0 \leq \omega$ ).

Let $\eta$ be the debt ratio in the economy ($\eta=\frac{D}{Y}$) (has units of years, because $D$ has units of dollars and $Y$ has units of dollars/year, $0 \leq \eta$).

The new variable is:

Let $p$ be the ratio of financial speculation to economic output ($p=\frac{P}{Y}$) (unitless).


### Model derivation:

#### Assumptions:

We change our assumption about debt. Debt can now be used for speculation.

$$ \frac{dD}{dt} = \kappa (\pi)Y - \pi Y + P $$ (+P is the new part)

#### Derivation:

The change in $P$ is an increasing function of the growth rate of the economy. The growth rate of the economy, as was derived for the Keen model, is 
$$\frac{dY/dt}{Y} = \frac{\kappa(1-\omega-r\eta)}{v} - \delta \equiv g(\lambda,\eta)$$

So $$\frac{dP}{dt}=\psi(g(\lambda,\eta))P$$

The growth rate of $p$ is

$$ \frac{dp/dt}{p}=\frac{d(P/Y)/dt}{P/Y} $$
$$ = \frac{dP/dt}{P} - \frac{dY/dt}{Y} $$
$$ = \psi(g(\lambda,\eta)) - g(\lambda,\eta) $$

Therefore,
$$ \frac{dp}{dt} = p[\psi(g(\lambda,\eta)) - g(\lambda,\eta)] $$

$$ \frac{dp}{dt} = p[\psi(\frac{\kappa(1-\omega-r\eta)}{v} - \delta) - \frac{\kappa(1-\omega-r\eta)}{v} + \delta] $$


The only other change that has to be made to the Keen model is to add p to the differential equation for debt ($\frac{d\eta}{dt}$), as debt can now be used for speculation. This gives Grasselli and Lima's model with financial speculation:

$$\frac{d\omega}{dt} = \omega [f(\lambda) - \alpha] $$

$$\frac{d\lambda}{dt} = \lambda [\frac{\kappa(1 -\omega - r\eta)}{v} - \delta - \alpha - \beta] $$

$$ \frac{d\eta}{dt} = \eta [r - \frac{\kappa(1-\omega-r\eta)}{v} + \delta] + \kappa(1-\omega-r\eta)- (1-\omega) + p$$

$$ \frac{dp}{dt} = p[\psi(\frac{\kappa(1-\omega-r\eta)}{v} - \delta) - \frac{\kappa(1-\omega-r\eta)}{v} + \delta] $$

### Analysis

#### Equilibria:

We do not analyse the stability of the 4D model. However, we note that if $p=0$, the 4D model collapse to the 3D model ($\frac{dp}{dt}=0$, $\frac{d\eta}{dt}$ is now the same as in the Keen model). This means that all equilibrium points from Keen hold for the 4D model if $p^{*}=0$.

### Numerical Analysis

#### Code for the model:

First we need a function for financial speculation. We choose a simple exponential function, $\psi(g(Y))=e^{20g(Y)-10}$, where $g(Y)$ is the growth rate of the economy.

```python
## defines and plots the speculation function for illustrative purposes

def spec(x,t,params):
    v, d = params
    return(np.exp(20*(x)-10))
xvec = np.arange(0,1,0.01)

a = spec(xvec,0,(3,0.02))

%matplotlib inline
plt.plot(xvec, a);
```

```python
#defines the 4D model, adds a variable for financial speculation
def ponzi(x, t, params):
    a, b, d, v, r  = params
    W, L, D, P = x
    return(np.array([W*(0.0001/((1-L)**2)-0.0204-a),
                    L*((np.exp(20*(1-W-r*D)-4.5)/v)-a-b-d),
                    D*(r-(np.exp(20*(1-W-r*D)-4.5)/v)+d)+np.exp(20*(1-W-r*D)-4.5)-(1-W)+P,
                    P*(np.exp(20*((np.exp(20*(1-W-r*D)-4.5)/v)+d)-10)-(np.exp(20*(1-W-r*D)-4.5)/v)+d)]))
```

#### Time-dependent solutions:

```python
tvec = np.arange(0,20,0.1)
attempt_1_4d = scipy.integrate.odeint(
    ponzi,
    y0 = (0.96,0.937,0.1,0.1),
    t = tvec,
    args = (params2,))
```

Therefore 4D model is sensitive to the initial conditions. The system appears to either goes to a positive limit cycle  or $(0,0,\infty,0)$. We create a diagram that shows which choice of initial conditions of $\lambda$ and $\omega$ causes the system to go to which equilibrium state.

```python
#### running system for various pairs values of omega and lambda, bruteforce
#### we are trying to see which starting conditions cause it to go to (0,0,inf,0)
#runs the system for starting values of lambda and omega
def runSystem(oga, lda,steps=300):
    tvec = np.arange(0,steps,0.1)
    output = scipy.integrate.odeint(
        ponzi,
        y0 = (oga, lda,0.1,0.1),
        t = tvec,
        args = (params2,))
    return output


step = 0.005

%matplotlib inline
plt.axes().set_aspect('equal', 'datalim')
plt.title("Starting Conditions Leading to Lambda and Omega = 0 Equilibrium")
plt.xlabel("$\omega_0$")
plt.ylabel("$\lambda_0$");

for oga in np.arange(0.7,.99,step):
    for lda in np.arange(0.7, 0.99, step):
        #get the final state of the system
        eqm = runSystem(oga,lda)[-1]
        
        #if omega and lambda tend to zero, mark this point red, as an undesirable equilibrium
        if 0.01 > eqm[0] and 0.01 > eqm[0]:
            plt.plot([oga],[lda],marker='o', markersize=1,color='red')
            
        #if eta tends to zero, mark this point black, as a desirable equilibrium
        if 0.01> eqm[3]:
            plt.plot([oga],[lda],marker='o', markersize=0.5,color='blue')
```

#### 4D model with income

Finally, we also add income to the 4D model.

```python
def ponzi_y(x, t, params):
    a, b, d, v, r  = params
    W, L, D, P, Y = x
    return(np.array([W*(0.0001/((1-L)**2)-0.0204-a),
                    L*((np.exp(20*(1-W-r*D)-4.5)/v)-a-b-d),
                    D*(r-(np.exp(20*(1-W-r*D)-4.5)/v)+d)+np.exp(20*(1-W-r*D)-4.5)-(1-W)+P,
                    P*(np.exp(20*((np.exp(20*(1-W-r*D)-4.5)/v)+d)-10)-(np.exp(20*(1-W-r*D)-4.5)/v)+d),
                     Y*((np.exp(20*(1-W-r*D)-4.5)/v)-d)]))
```

```python
# find time dependent solution of 4D with income, plot with two different scales and a legend
tvec = np.arange(0,100,0.1)
attempt_y_4d = scipy.integrate.odeint(
    ponzi_y,
    y0 = (0.96,0.9,0.1,0.1,100),
    t = tvec,
    args = (params2,))

fig, ax1 = plt.subplots()
ax1.plot(tvec, attempt_y_4d[:,0], 'steelblue')
ax1.plot(tvec, attempt_y_4d[:,1], 'darkorange')
ax1.plot(tvec, attempt_y_4d[:,2], 'forestgreen')
ax1.plot(tvec, attempt_y_4d[:,3], 'orchid')
ax1.set_xlabel('time')
ax1.set_ylabel('$\lambda$, $\omega$, $\eta$, p')

ax2 = ax1.twinx()
ax2.plot(tvec, attempt_y_4d[:,4], 'indianred')
ax2.set_ylabel('Y')

blue_patch = mpatches.Patch(color='steelblue', label='$\omega$')
orange_patch = mpatches.Patch(color='darkorange', label='$\lambda$')
green_patch = mpatches.Patch(color='forestgreen', label='$\eta$')
pink_patch = mpatches.Patch(color='orchid', label='p')
red_patch = mpatches.Patch(color='indianred', label='Y')
plt.legend(handles=[blue_patch,orange_patch,green_patch,pink_patch,red_patch])

plt.show()
```

## Conclusion

We have seen three models of the economy. The first model, the 2D 'Goodwin' model, was a variation of the Lotka-Volterra predator-prey model, with $\lambda$ in the role of prey and $\omega$ in the role of predator. In this model, the economy grows over time, with some deviations from trend, as in real life. However, the model does not allow for a financial crisis because the nonzero equilibrium is stable while the zero equilibrium is stable, and this is independent of any  choice of starting conditions or any reasonable choice of paramters. We have this simple result because the model does not allow for borrowing or speculation, so ignores the role of the financial sector in the economy. As we know from the 2007-09 recession, the health of the financial sector can affect the health of the overall economy. For this reason, while the Goodwin model is a good starting point, it is not complex enough to allow for reasonable economic outcomes.

The second model, the 3D 'Keen' model, introduces borrowing, and hence the financial sector, into the model economy. However, debt can only be used to finance investment, not financial speculation. The model becomes more complex, but there are still two equilibrium points (the third only exists for a particular set of paramaters). The nonzero equilibrium remains stable, while the equilibrium point with $\omega=\lambda=0$ is unstable. This model still does not allow for the type of economic outcome we saw in 2007-09, as debt can only be used to create capital.

The third model is the 4D model from Grasselli and Lima (2012), which we call the 'Financial Speculation' model. Their model allows for financial speculation, an important change from the Keen model. While we do not examine this model as thoroughly as the other two, we show that the stability of this model is sensitive to both initial conditions and the choice of parameters. This added complexity means that a change in the underlying economic conditions (the paramters or the speculation function) could destabilize the nonzero equilibrium, causing the path of income to deviate from its trend. This model is the most realistic, as we expect a change in economic fundamentals in real life could knock the growth rate of the economy from its upward trend.
